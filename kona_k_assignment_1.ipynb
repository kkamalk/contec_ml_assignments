```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt

```

```python
plt.rc('font',size=12)
plt.rc('axes',labelsize=14,titlesize=14)
```

```python
df = pd.read_csv("/content/king_county_house_data (1).csv")
```

```python
df.head()
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
X=df[['floors']].values
y=df[['price']].values
```

```python
# Create a scatter plot of price vs bedrooms
df.plot(kind='scatter', grid=True, x='floors', y='price')
plt.show();





```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
print(model.coef_)
```

```python
y_pred=model.predict(X)
```

```python
y_pred
```

```python
score = model.score(X, y)
print('R-squared score:', score)
```

```python
results = pd.DataFrame({'Actual': y.flatten(), 'Predicted': y_pred.flatten()})
print(results.head())
```
